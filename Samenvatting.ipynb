# Multivalued Dependencies (MVD's) en 4NF

## Definitie van een Multivalued Dependency (MVD)

Zij $R$ een relatie en $X, Y \subseteq R$.

We zeggen dat er een **multivalued dependency** geldt:

$$
X \twoheadrightarrow Y
$$

als voor alle tuples $t_1, t_2 \in R$ waarvoor geldt:

$$
t_1[X] = t_2[X],
$$

er ook tuples $t_3, t_4 \in R$ bestaan zodanig dat:

$$
t_3[X] = t_4[X] = t_1[X],
$$

$$
t_3[Y] = t_1[Y],
$$

$$
t_3[R - (X \cup Y)] = t_2[R - (X \cup Y)],
$$

$$
t_4[Y] = t_2[Y],
$$

$$
t_4[R - (X \cup Y)] = t_1[R - (X \cup Y)].
$$

### Intuïtieve betekenis

Een MVD $X \twoheadrightarrow Y$ betekent:

Gegeven een vaste waarde van $X$, zijn de waarden van $Y$ onafhankelijk van de overige attributen.

Met andere woorden: voor één $X$ krijg je een cartesisch product tussen $Y$ en de rest van de attributen.

### Relatie met Functionele Afhankelijkheid (FD)

Een functionele afhankelijkheid:

$$
X \rightarrow Y
$$

impliceert altijd:

$$
X \twoheadrightarrow Y
$$

Elke FD is dus een speciale vorm van een MVD.

# Exercise 20

We hebben relatie:

$$
ANIMAL(\text{lname}, \text{species}, \text{food}, \text{continent})
$$

waar:

- $\text{lname}$ = Latijnse naam (identificeert elke soort)
- $\text{species}$ = soortnaam
- $\text{food}$ = type voedsel
- $\text{continent}$ = continent waar de soort voorkomt

---

## Stap 1 — Functionele Afhankelijkheden

Uit de beschrijving volgt:

$$
\text{lname} \rightarrow \text{species}
$$

Dit is de enige niet-triviale FD.

---

## Stap 2 — Multivalued Dependencies

We maken de aanname:

Het dieet van een soort is onafhankelijk van de continenten waarop de soort voorkomt.

Dus:

$$
\text{food} \perp \text{continent} \mid \text{lname}
$$

Daaruit volgen de MVD's:

$$
\text{lname} \twoheadrightarrow \text{food}
$$

$$
\text{lname} \twoheadrightarrow \text{continent}
$$

---

## Stap 3 — Sleutel

Omdat een soort meerdere soorten voedsel en meerdere continenten kan hebben, is de sleutel:

$$
(\text{lname}, \text{food}, \text{continent})
$$

---

## Stap 4 — 4NF Controle

Een relatie zit in 4NF als:

Voor elke niet-triviale MVD $X \twoheadrightarrow Y$ is $X$ een superkey.

Hier geldt:

$$
\text{lname} \twoheadrightarrow \text{food}
$$

maar $\text{lname}$ is geen superkey.

Dus:

$$
ANIMAL \text{ is niet in 4NF.}
$$

---

## Stap 5 — 4NF Decompositie

We decomponeren langs de MVD's.

Eindresultaat:

$$
ANIMAL_1(\text{lname}, \text{species})
$$

$$
ANIMAL_2(\text{lname}, \text{food})
$$

$$
ANIMAL_3(\text{lname}, \text{continent})
$$

Elke relatie zit nu in 4NF.

---

## Samenvatting

FD:
$$
\text{lname} \rightarrow \text{species}
$$

MVD's (onder onafhankelijkheidsassumptie):
$$
\text{lname} \twoheadrightarrow \text{food}
$$
$$
\text{lname} \twoheadrightarrow \text{continent}
$$

4NF-decompositie:
$$
(\text{lname}, \text{species}),\;
(\text{lname}, \text{food}),\;
(\text{lname}, \text{continent})
$$

Een MVD zegt eigenlijk, twee attributen kunnen onafhankelijk van elkaar meerder waarden aannemen, maar dat moeten we alle mogelijke combinaties in de relatie zitten.
