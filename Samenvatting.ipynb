# Multivalued Dependencies (MVD's) en 4NF

## Definitie van een Multivalued Dependency (MVD)

Zij $R$ een relatie en $X, Y \subseteq R$.

We zeggen dat er een **multivalued dependency** geldt:

$$
X \twoheadrightarrow Y
$$

als voor alle tuples $t_1, t_2 \in R$ waarvoor geldt:

$$
t_1[X] = t_2[X],
$$

er ook tuples $t_3, t_4 \in R$ bestaan zodanig dat:

$$
t_3[X] = t_4[X] = t_1[X],
$$

$$
t_3[Y] = t_1[Y],
$$

$$
t_3[R - (X \cup Y)] = t_2[R - (X \cup Y)],
$$

$$
t_4[Y] = t_2[Y],
$$

$$
t_4[R - (X \cup Y)] = t_1[R - (X \cup Y)].
$$

### IntuÃ¯tieve betekenis

Een MVD $X \twoheadrightarrow Y$ betekent:

Gegeven een vaste waarde van $X$, zijn de waarden van $Y$ onafhankelijk van de overige attributen.

Met andere woorden: voor Ã©Ã©n $X$ krijg je een cartesisch product tussen $Y$ en de rest van de attributen.

### Relatie met Functionele Afhankelijkheid (FD)

Een functionele afhankelijkheid:

$$
X \rightarrow Y
$$

impliceert altijd:

$$
X \twoheadrightarrow Y
$$

Elke FD is dus een speciale vorm van een MVD.

# Exercise 20

We hebben relatie:

$$
ANIMAL(\text{lname}, \text{species}, \text{food}, \text{continent})
$$

waar:

- $\text{lname}$ = Latijnse naam (identificeert elke soort)
- $\text{species}$ = soortnaam
- $\text{food}$ = type voedsel
- $\text{continent}$ = continent waar de soort voorkomt

---

## Stap 1 â€” Functionele Afhankelijkheden

Uit de beschrijving volgt:

$$
\text{lname} \rightarrow \text{species}
$$

Dit is de enige niet-triviale FD.

---

## Stap 2 â€” Multivalued Dependencies

We maken de aanname:

Het dieet van een soort is onafhankelijk van de continenten waarop de soort voorkomt.

Dus:

$$
\text{food} \perp \text{continent} \mid \text{lname}
$$

Daaruit volgen de MVD's:

$$
\text{lname} \twoheadrightarrow \text{food}
$$

$$
\text{lname} \twoheadrightarrow \text{continent}
$$

---

## Stap 3 â€” Sleutel

Omdat een soort meerdere soorten voedsel en meerdere continenten kan hebben, is de sleutel:

$$
(\text{lname}, \text{food}, \text{continent})
$$

---

## Stap 4 â€” 4NF Controle

Een relatie zit in 4NF als:

Voor elke niet-triviale MVD $X \twoheadrightarrow Y$ is $X$ een superkey.

Hier geldt:

$$
\text{lname} \twoheadrightarrow \text{food}
$$

maar $\text{lname}$ is geen superkey.

Dus:

$$
ANIMAL \text{ is niet in 4NF.}
$$

---

## Stap 5 â€” 4NF Decompositie

We decomponeren langs de MVD's.

Eindresultaat:

$$
ANIMAL_1(\text{lname}, \text{species})
$$

$$
ANIMAL_2(\text{lname}, \text{food})
$$

$$
ANIMAL_3(\text{lname}, \text{continent})
$$

Elke relatie zit nu in 4NF.

---

## Samenvatting

FD:
$$
\text{lname} \rightarrow \text{species}
$$

MVD's (onder onafhankelijkheidsassumptie):
$$
\text{lname} \twoheadrightarrow \text{food}
$$
$$
\text{lname} \twoheadrightarrow \text{continent}
$$

4NF-decompositie:
$$
(\text{lname}, \text{species}),\;
(\text{lname}, \text{food}),\;
(\text{lname}, \text{continent})
$$

Een MVD zegt eigenlijk, twee attributen kunnen onafhankelijk van elkaar meerder waarden aannemen, maar dat moeten we alle mogelijke combinaties in de relatie zitten.

# Analyse van de Trigger NetWorth

$$
\textbf{Gegeven trigger}
$$

$$
\begin{aligned}
&\text{CREATE TRIGGER NetWorth} \\
&\text{AFTER UPDATE OF netWorth ON FilmProd} \\
&\text{REFERENCING} \\
&\text{OLD ROW AS OldTuple} \\
&\text{NEW ROW AS NewTuple} \\
&\text{FOR EACH ROW} \\
&\text{WHEN } (OldTuple.netWorth > NewTuple.netWorth) \\
&\text{UPDATE FilmProd SET netWorth = OldTuple.netWorth}
\end{aligned}
$$

---

$$
\textbf{1. Wat doet deze trigger?}
$$

De trigger wordt uitgevoerd wanneer:

$$
\text{Event: } UPDATE \text{ op attribuut } netWorth
$$

Voor elke gewijzigde tuple geldt:

$$
t_{\text{old}} = OldTuple
\qquad
t_{\text{new}} = NewTuple
$$

De voorwaarde is:

$$
OldTuple.netWorth > NewTuple.netWorth
$$

Dat betekent:

$$
netWorth_{\text{old}} > netWorth_{\text{new}}
$$

Dus de trigger wordt geactiveerd wanneer het vermogen
wordt verlaagd.

Wanneer de voorwaarde waar is, wordt uitgevoerd:

$$
netWorth := netWorth_{\text{old}}
$$

Formeel betekent dit:

$$
(netWorth_{\text{old}} > netWorth_{\text{new}})
\Rightarrow
netWorth_{\text{new}} := netWorth_{\text{old}}
$$

---

$$
\textbf{2. Conceptuele betekenis}
$$

De trigger forceert de invariant:

$$
netWorth_{\text{new}} \ge netWorth_{\text{old}}
$$

Met andere woorden:

$$
\text{Het vermogen mag nooit dalen.}
$$

Elke poging tot verlaging wordt onmiddellijk teruggezet.

---

$$
\textbf{3. Interne werking}
$$

Bij een statement:

$$
UPDATE\ FilmProd\ SET\ netWorth = w_{\text{new}}
$$

voert de database-engine uit:

$$
\begin{aligned}
1.\;& t \in FilmProd \text{ wordt geselecteerd} \\
2.\;& t_{\text{old}} := t \\
3.\;& t_{\text{new}} := t \text{ met } netWorth = w_{\text{new}} \\
4.\;& \text{Controleer } w_{\text{old}} > w_{\text{new}} \\
5.\;& \text{Indien waar: herstel } w_{\text{new}} := w_{\text{old}}
\end{aligned}
$$

---

$$
\textbf{4. Algemene definitie van een row-level trigger}
$$

Een trigger kan formeel worden gedefinieerd als:

$$
f : (OLD, NEW) \rightarrow \text{side effects}
$$

Algemene syntactische vorm:

$$
\begin{aligned}
&\text{CREATE TRIGGER } T \\
&\{ \text{BEFORE} \mid \text{AFTER} \}
\{ \text{INSERT} \mid \text{UPDATE} \mid \text{DELETE} \} \\
&\text{ON } R \\
&\text{FOR EACH ROW} \\
&\text{WHEN } (\varphi(OLD,NEW)) \\
&\text{BEGIN statements END}
\end{aligned}
$$

---

$$
\textbf{5. Wiskundige interpretatie}
$$

Een row-trigger bewaakt een invariant:

$$
I(t)
$$

Bij een tuple-transformatie:

$$
t_{\text{old}} \longrightarrow t_{\text{new}}
$$

De trigger zorgt ervoor dat:

$$
I(t_{\text{new}})
$$

waar blijft na elke update.

In dit geval is de invariant:

$$
netWorth_{\text{new}} \ge netWorth_{\text{old}}
$$

### Statement-level trigger voor Maxsal

Gegeven:

- EMP(empno, name, sal)  
- Maxsal(val) bevat exact Ã©Ã©n tuple

### Trigger

```sql
CREATE TRIGGER SALARY
AFTER UPDATE OF sal ON EMP
FOR EACH STATEMENT
UPDATE Maxsal
SET val = (SELECT MAX(sal) FROM EMP);
```

---

### Werking

Een statement-level trigger wordt exact **Ã©Ã©n keer per SQL-statement** uitgevoerd,
ongeacht hoeveel tuples gewijzigd worden.

Bij:

```sql
UPDATE EMP SET sal = sal + 100;
```

gebeurt:

1. Alle betreffende tuples in EMP worden aangepast.
2. Daarna wordt de trigger precies Ã©Ã©n keer uitgevoerd.
3. De maximale salariswaarde wordt opnieuw berekend.

---

### Invariant

De trigger bewaart de invariant:

$$
\text{Maxsal.val} = \max(\text{EMP.sal})
$$

Na elke update geldt dus opnieuw:

$$
\text{Maxsal} := \max(\text{EMP})
$$

---

### Conceptueel verschil

Row-level trigger:

$$
\forall t \in \Delta EMP
$$

(incrementeel onderhoud per tuple)

Statement-level trigger:

$$
\text{na uitvoering van } \Delta EMP
$$

(volledige herberekening van de aggregatie)

---

### Interpretatie

Row-level = incremental view maintenance  
Statement-level = full view recomputation
