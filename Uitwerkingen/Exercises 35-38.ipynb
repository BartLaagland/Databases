#### 36.

**a**

We weten dat vóór de commit de log naar de disk geschreven moet worden, na de commit mag de dirty data naar de disk worden geschreven, dit kan echter ook later. Hierdoor kan je dus in een keer veel data overschrijven naar de disk, dit is efficiënter. Als er een crash ontstaat tijdens het schrijven van een log dan zal het systeem denken dat er nooit een verandering plaats gevonden heeft, aangezien de log dan niet op de disk staat en er niks is veranderd qua data op de disk. Mocht er een crash zijn na het schrijven van de log en voor of tijdens het overschrijven van de data kun je door middel van REDO dus de veranderingen alsnog toepassen in de data base. 

**b**

Nee, er kunnen ook nog veranderingen zijn die gedeeltelijk hebben plaatsgevonden. Een REDO actie garandeerd dat alle committed transacties opnieuw worden uitgevoerd waarbij je zeker weet dat die transacties up to date zijn. Echter hebben we ook de notie uncommited transacties, die moeten nog ongedaan worden (UNDO).

#### 37.

Wanneer we UNDO en REDO gaan combineren willen we eerst de REDO doen waarbij we in chronologische volgorde werken. De eerste committed transactie (tijdsgewijs) wordt dus opnieuw uitgevoerd tot en met de laatste committed transactie. Waarbij de eerste committed transactie verwijst naar de eerste commit na een de laatste checkpoint. Daarna doen we UNDO juist in de tegengestelde richting, we maken dus de laatste uncommitted transactie ongedaan tot en met de eerste uncommitted transactie, nogmaals met respect naar de laatste checkpoint. 

Met betrekking tot de vrijheid van de buffer manager zien we dus dat de buffer manager zelf kan bepalen wanneer hij de data flushed. Zij het bij iedere commit, zij het om de 2 uur. Met dit recovery process zal het altijd goed komen bij een crash. Als de buffer manager zich maar aan de regel houd dat na commit de data overgeschreven **mag** worden, dit kan dus ook veel later na de commit.

#### 38.

Wanneer er een kopie van de database op een aparte disk (backup) wordt opgeslagen, en de log vanaf dat moment ook op een medium wordt bewaard dat niet verloren gaat bij een crash van de data disk, kan men het media failure probleem oplossen met behulp van logging.

Als de data disk crasht of kapot gaat, kan eerst de backup van de database worden teruggezet. Vervolgens kan men de log gebruiken om alle wijzigingen die na het moment van de backup hebben plaatsgevonden opnieuw toe te passen.

Omdat de backup een consistente toestand van de database bevat, hoeven enkel de committed updates na de backup opnieuw uitgevoerd te worden, we passen dus alleen REDO toe

Het belangrijkste vereiste is dus dat de **log overleeft wanneer de data disk crasht**, zodat de updates na de backup opnieuw kunnen worden toegepast op de herstelde database.
