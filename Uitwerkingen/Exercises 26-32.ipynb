We zagen dat $S1$ niet serializable is, en kan hij dus ook niet 2PL zijn aangezien, 2PL $\Rightarrow$ Serializable. We bekijken $S2$ en herinneren de conflicts:

$$
    T_5 \rightarrow T_4, \\
    T_5 \rightarrow T_2, \\
    T_3 \rightarrow T_2, \\
    T_1 \rightarrow T_3, \\
    T_1 \rightarrow T_5, \\
    T_4 \rightarrow T_2, \\
    T_3 \rightarrow T_5
$$

We volgen de compatibility matrix voor locks:

$$
\begin{array}{c|cc}
 & R & W \\
\hline
R & \text{yes} & \text{no} \\
W & \text{no} & \text{no}
\end{array}
$$

We zien dat op $z$ eerst een $W$ lock komt en dan een $R$ lock, dit is niet mogelijk.

#### 32.

Gebruik van wait-for graphs komt er op neer dat we alle conflicten met betrekking tot transacties, die wachten op een data type vanwege een lock, modelleren in een graaf waarbij de nodes transacties zijn en de edges de conflicts. Mocht deze graaf cyclisch zijn dan is er een deadlock en moeten we het schema dus aanpassen!
