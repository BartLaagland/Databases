#### 39.

Een netwerk partitie wil zeggen, een situatie waarin een gedistribueerd systeem (database in dit geval) wordt opgesplitst in meerdere groepen van nodes die onderling niet meer kunnen communiceren. Dus wanneer er in bijvoorbeeld groep $A$ en groep $B$ het CTP wordt gestart, dan kan dat resulteren in inconsistenties. We werken een volldig voorbeeld uit:

**Groep $A$ start CTP**

- De COORD is om een of andere reden gefaald, CTP wordt gestart
- De algemene beslissing $\langle \text{COMMIT } T \rangle$ is genomen
- Tenminste één PART heeft deze globale commit ontvangen
- de PART die dit heeft ontvangen broadcast dit naar de rest als COORD
- Groep $A$ commit

**Groep $B$ start CTP**

- De COORD is om een of andere reden gefaald, CTP wordt gestart
- Geen enkele PART kent de globale beslissing (geen $\langle \text{COMMIT } T \rangle$ of $\langle \text{ABORT } T \rangle$)  
- Er bestaat een PART zonder $\langle \text{READY } T \rangle$, dus deze heeft nog niet YES gestemd  
- De nieuwe COORD kan daarom veilig een $\langle\text{GLOBAL ABORT}\rangle$ beslissen  
- Groep $B$ abort  

Dit resulteert er dus in dat in een gedistribueerd systeem twee verschillende versies bestaan tegelijkertijd
$$
 \Rightarrow \quad \text{inconsistent.}
$$

#### 40.

Laat $A$ de PART in kweste zijn en $B$ de bijbehorende COORD. We werken het scenario uit waarin deze gedachte gang logisch, en dus ook nodig is.

- $B$ stuurt een $\langle \text{PREPARE } T \rangle$
- $A$ ontvangt $\langle \text{PREPARE } T \rangle$
- $A$ checkt of zij kan committen, antwoord is ja
- $A$ logt $\langle \text{READY } T \rangle$
- $A$ crasht
- $A$ heeft de stem nog niet gestuurd naar $B$
- $A$ bekijkt de log en ziet dat zij instaat is om te committen, zij stuurt $\langle \text{READY } T \rangle$ naar $B$
- $B$ ontvangt $\langle \text{READY } T \rangle$ van $A$ en beslist dan of iedereen moet committen of niet

Had $A$ niet gelogt dat zij ready was had zij na de crash dus niet $\langle \text{READY } T \rangle$ gestuurd en was er een global abort ontstaan, wat onnodig was. Dit is ook gelijk de reden dat bij een $\langle \text{ABORT } T \rangle$ dit niet gelogt hoeft te worden, bekijk het analoge voorbeeld hieronder

- $B$ stuurt een $\langle \text{PREPARE } T \rangle$
- $A$ ontvangt $\langle \text{PREPARE } T \rangle$
- $A$ checkt of zij kan committen, antwoord is $\textbf{nee}$
- $A$ logt $\langle \text{ABORT } T \rangle$ dus $\textbf{NIET}$
- $A$ crasht
- $A$ heeft de stem nog niet gestuurd naar $B$
- $A$ bekijkt de log en ziet niks staan, veiligheidshalve stuurt $A$ $\langle \text{ABORT } T \rangle$ naar $B$
- $B$ ontvangt $\langle \text{ABORT } T \rangle$ van $A$ en beslist een global abort
