**Definities**

$\textbf{Domein}$

Een domein $D$ is een verzameling van atomaire waarden. 
(Bijvoorbeeld: gehele getallen, tekens, strings, floats, geldbedragen, datums)

$\textbf{Relatieschema}$

Een relatieschema $R$, genoteerd als $R(A_1, A_2, \dots, A_n)$, bestaat uit een relatienaam $R$ en een verzameling attributen $A_1, A_2, \dots, A_n$; 
$n$ is de graad van het schema $R$.

We definiëren $\mathrm{attr}(R) = \{A_1, A_2, \dots, A_n\}$.

$\textbf{Relatie}$

Een relatie $r$ over een schema $R$, genoteerd als $r(R)$, is een verzameling van tupels 
$\langle v_1, v_2, \dots, v_n \rangle$, waarbij elke $v_i \in D_i \cup \{\text{null}\}$.
Dus elk attribuut $A_i$ is gekoppeld aan een specifiek domein $D_i$.


$\textbf{Superkey}$

Een verzameling attributen $K \subseteq \mathrm{attr}(R)$ is een superkey als voor elke geldige relatie $r(R)$ geldt:  
$t_1, t_2 \in r : t_1[K] = t_2[K] \Rightarrow t_1 = t_2$

$\textbf{Candidate key}$

Een superkey $K \subseteq \mathrm{attr}(R)$ is een candidate key als er geen $K' \subset K$, $K' \neq K$, bestaat die ook een superkey is.

$\textbf{Primary key}$

Eén van de candidate keys wordt (om een bepaalde reden) gekozen als primary key.

$\textbf{Foreign key}$

Een foreign key is een verzameling attributen $K \subseteq \mathrm{attr}(R_i)$ die voorkomt in een andere relatie $R_j$ als een candidate key.  
We zeggen dat $R_i[K]$ verwijst naar $R_j[K]$.

$\textbf{Selection operator } \sigma$

De selection operator $\sigma_p(R)$ selecteert alle tupels uit relatie $R$ die voldoen aan predicaat $p$.

Formeel:  
$\sigma_p(R) = \{\, t \in R \mid p(t) \,\}$

$\textbf{Projection operator } \pi$

De projection operator $\pi_L(R)$ selecteert uit relatie $R$ alleen de attributen in lijst $L$.

Formeel:  
$\pi_L(R) = \{\, t[L] \mid t \in R \,\}$

$\textbf{Union operator } \cup$

De union operator $R \cup S$ geeft alle tupels die in $R$ of in $S$ voorkomen (zonder duplicaten), op voorwaarde dat $R$ en $S$ schema-compatibel zijn.

Formeel:  
$R \cup S = \{\, t \mid t \in R \lor t \in S \,\}$

$\textbf{Difference operator } -$

De difference operator $R - S$ geeft alle tupels die in $R$ voorkomen maar niet in $S$, op voorwaarde dat $R$ en $S$ schema-compatibel zijn.

Formeel:  
$R - S = \{\, t \mid t \in R \land t \notin S \,\}$

$\textbf{Intersection operator } \cap$

De intersection operator $R \cap S$ geeft alle tupels die zowel in $R$ als in $S$ voorkomen, op voorwaarde dat $R$ en $S$ schema-compatibel zijn.

Formeel:  
$R \cap S = \{\, t \mid t \in R \land t \in S \,\}$

$\textbf{Cartesian product } \times$

Het Cartesian product $R \times S$ geeft alle mogelijke combinaties van tupels uit $R$ en $S$.

Formeel:  
$R \times S = \{\, (t, u) \mid t \in R \land u \in S \,\}$

$\textbf{Theta-join } \bowtie_\theta$

De theta-join $R \bowtie_\theta S$ geeft alle combinaties van tupels uit $R$ en $S$ die voldoen aan conditie $\theta$.

Formeel:  
$R \bowtie_\theta S = \{\, (t, u) \mid t \in R \land u \in S \land \theta(t, u) \,\}$

$\textbf{Natural join } \bowtie$

De natural join $R \bowtie S$ geeft alle combinaties van tupels uit $R$ en $S$ waarbij gemeenschappelijke attributen dezelfde waarde hebben.

Formeel:  
$R \bowtie S = \{\, (t, u) \mid t \in R \land u \in S \land t[A] = u[A] \text{ voor alle gemeenschappelijke attributen } A \,\}$

$\textbf{Division operator } \div$

De division operator $T \div U$ geeft alle waarden van attributen in $T$ die gecombineerd voorkomen met alle waarden in $U$.

Formeel:  
$T \div U = \{\, t[A] \mid t \in T \land \forall u \in U : (t[A], u) \in T \,\}$

$\textbf{Assignment } :=$

De assignment operator $S := R$ kent de uitkomst van expressie $R$ toe aan naam $S$.

Formeel:  
$S := R$

$\textbf{Renaming operator } \rho$

De renaming operator $\rho_{S(A_1,\dots,A_n)}(R)$ hernoemt relatie $R$ naar $S$ en/of hernoemt de attributen naar $A_1,\dots,A_n$.

Formeel:  
$\rho_{S(A_1,\dots,A_n)}(R) = \{\, (A_1:=v_1,\dots,A_n:=v_n) \mid (v_1,\dots,v_n) \in R \,\}$

$\textbf{Ordering operator } \tau$

De ordering operator $\tau_{L}(R)$ sorteert de tupels van $R$ volgens attributen in lijst $L$.

Formeel:  
$\tau_{L}(R) = \text{gesorteerde volgorde van } R \text{ volgens } L$

$\textbf{Grouping en aggregatie } \Gamma$

De grouping operator $\Gamma_{G,\,f(A)}(R)$ groepeert de tupels van $R$ volgens attributen $G$ en past een aggregatiefunctie $f$ toe op attribuut $A$ per groep.

Formeel:  
$\Gamma_{G,\,f(A) \rightarrow B}(R) = \{\, (g,\, f(\{\, t[A] \mid t \in R \land t[G]=g \,\})) \mid g \in \pi_G(R) \,\}$

**Stellingen**

$\textbf{Key integrity constraint}$

De key integrity constraint verbiedt $\text{null}$-waarden in een primary key kolom.

$\textbf{Referential integrity constraint}$

De referential integrity constraint stelt dat:  
als $R_i[K]$ een foreign key is die verwijst naar $R_j[K]$,  
dan geldt $R_i[K] \subseteq R_j[K] \cup \{\text{null}\}$.

$\textbf{Cascading en commuting selections}$

$\sigma_p(\sigma_q(R)) = \sigma_q(\sigma_p(R)) = \sigma_{p \land q}(R)$

$\textbf{Cascading projections}$

Zij $L_1 \subseteq L_2$, dan geldt

$\pi_{L_1}(\pi_{L_2}(R)) = \pi_{L_1}(R)$ 

$\textbf{Commuting selections en projections}$

Zij $\mathrm{attr}(p) \subseteq L$, dan geldt

$\pi_L(\sigma_p(R)) = \sigma_p(\pi_L(R))$

$\textbf{Commutativiteit}$

$R \cup S = S \cup R$

$R \cap S = S \cap R$

$R \bowtie S = S \bowtie R$

$\textbf{Associativiteit}$

$(R \cup S) \cup T = R \cup (S \cup T)$

$(R \cap S) \cap T = R \cap (S \cap T)$

$(R \bowtie S) \bowtie T = R \bowtie (S \bowtie T)$

$\textbf{Distributiviteit van } \sigma$

$\sigma_p(R \cup S) = \sigma_p(R) \cup \sigma_p(S)$

$\textbf{Stelling (Distributiviteit van selectie over join)}$

Zij $\mathrm{attr}(p_1) \subseteq \mathrm{attr}(R)$ en $\mathrm{attr}(p_2) \subseteq \mathrm{attr}(S)$, dan geldt

$\sigma_{p_1 \land p_2 \land p_3}(R \bowtie S) 
= \sigma_{p_3}(\sigma_{p_1}(R) \bowtie \sigma_{p_2}(S))$

**Voorbeelden van complexe RA operatoren en haar respectievelijke predicaten**

$\textbf{Extended projection } \pi$

De extended projection operator $\pi_{L}(R)$ laat toe dat attributen in $L$ ook expressies zijn.

Formeel:  
$\pi_{e_1 \rightarrow A_1,\dots,e_k \rightarrow A_k}(R) = \{\, (A_1 := e_1(t),\dots,A_k := e_k(t)) \mid t \in R \,\}$

Voorbeeld:

$$
R =
\begin{array}{c|c}
A & B \\
\hline
1 & 4 \\
2 & 5 \\
3 & 6
\end{array}
\qquad
\pi_{A,\,A+B}(R) =
\begin{array}{c|c}
A & A + B \\
\hline
1 & 5 \\
2 & 7 \\
3 & 9
\end{array}
$$

$\textbf{Grouping } \Gamma$

Voorbeeld:

$$
S =
\begin{array}{c|c|c}
Store & Product & Price \\
\hline
AH & Tomato\ polpa & 2.25 \\
Jumbo & Linguine & 2.52 \\
Aldi & Tomato\ polpa & 1.29 \\
AH & Mozzarella & 1.95 \\
Jumbo & Tomato\ polpa & 2.25 \\
AH & Linguine & 2.79
\end{array}
 \quad
\Gamma_{Product,\,MIN(Price)}(S) =
\begin{array}{c|c}
Product & MinPrice \\
\hline
Tomato\ polpa & 1.29 \\
Linguine & 2.52 \\
Mozzarella & 1.95
\end{array}
$$

Hier is product dus de attribuut en $MIN(price)$ de aggregate functie. We groeperen dus eerste alle producten, en pakken voor elk producten (type) het product gekoppeld aan de laagste prijs.

$\textbf{Associativiteit en join-volgorde}$

Door de associativiteit van de join-operator geldt:
$(R \bowtie S) \bowtie T = R \bowtie (S \bowtie T)$

Hierdoor bestaan er veel mogelijke manieren om een join-keten te evalueren, bv.:
$R \bowtie S \bowtie T \bowtie U$

kan op verschillende manieren geparentheseerd worden, zoals:
$((R \bowtie S) \bowtie T) \bowtie U$,  
$R \bowtie (S \bowtie (T \bowtie U))$,  
$(R \bowtie S) \bowtie (T \bowtie U)$, enz.

$\textbf{Belangrijk inzicht}$

Het aantal mogelijke join-volgordes groeit exponentieel met het aantal relaties $n$.

$\textbf{Query optimalisatie}$

Het kiezen van de beste join-volgorde is cruciaal voor performance:

- Voor kleine $n$: optimale oplossing via dynamic programming  
- Voor grote $n$: gebruik van heuristieken  

$\textbf{Kost van een join}$

De grootte van een relatie wordt typisch benaderd door:
aantal tupels $\times$ breedte (in bytes)

Deze grootte beïnvloedt de kost van volgende joins.

$\textbf{Heuristiek (greedy)}$

Een veelgebruikte strategie:
- schat de grootte van elke mogelijke join  
- kies telkens de kleinste intermediate resultaten  
- herhaal iteratief

$\textbf{Belangrijke opmerking}$

Een natural join reduceert tot een Cartesian product als er geen gemeenschappelijke attributen zijn.
