# 3NF Decomposition Using a Minimal Cover
This notebook summarizes the **3NF decomposition algorithm** based on a **minimal cover**.
We illustrate every step using the example:

**Relation:**
R(A,B,C,D,E,G)

**Functional Dependencies:**
BD → CEG,  D → BE,  E → G


## Step 1 — Split Right-Hand Sides
Each functional dependency must have **one attribute on the right side**.

BD → CEG becomes:
- BD → C
- BD → E
- BD → G

So we obtain:

F' = {BD → C, BD → E, BD → G, D → B, D → E, E → G}


## Step 2 — Reduce Left Sides
We check if attributes on the **left side** are redundant.

Because **D → B**, the attribute **B is redundant** in dependencies starting with BD.

Thus:
- BD → C becomes **D → C**
- BD → E becomes **D → E**
- BD → G becomes **D → G**

Result:

F'' = {D → C, D → E, D → G, D → B, D → E, E → G}


## Step 3 — Remove Redundant Dependencies
Now we remove FDs that can be derived from others.

- Duplicate: D → E appears twice → remove one.
- Because D → E and E → G, we can derive D → G.
  Therefore **D → G is redundant**.

Minimal cover:

F_min = {D → C, D → B, D → E, E → G}


## Step 4 — Create Relations
For each FD X → Y create relation **X ∪ Y**.

- D → C  →  R₁(D,C)
- D → B  →  R₂(D,B)
- D → E  →  R₃(D,E)
- E → G  →  R₄(E,G)


## Step 5 — Ensure a Key Exists
We must check if one of the relations contains a **candidate key of R**.

Compute closure of D:

D⁺ = {B,C,D,E,G}

Attribute **A** is missing, so D is not a key.

Candidate key:

AD


## Step 6 — Guarantee Lossless Join
If none of the relations contains a key, we **add one relation containing a key**.

Add:

R₅(A,D)


## Final 3NF Decomposition

R₁(D,C)

R₂(D,B)

R₃(D,E)

R₄(E,G)

R₅(A,D)

This decomposition:
- Preserves dependencies
- Is in **Third Normal Form (3NF)**
- Guarantees a **lossless join**


## Quick Exam Strategy
1. Split RHS attributes.
2. Reduce left sides.
3. Remove redundant dependencies.
4. Create a relation for each FD.
5. Check if a candidate key appears in a relation.
6. If not, add one relation containing a key.
